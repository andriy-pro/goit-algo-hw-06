# Порівняння способів розрахунку у Python відстані між точками на поверхні Землі

Розрахунок відстані можливий за допомогою різних підходів, що базуються на використанні географічних координат (широта і довгота). У цій роботі були застосовані наступні методи:

1. **Бібліотека `haversine`**  
   Надає можливість обчислення відстаней між точками на поверхні Землі за допомогою [формули гаверсинуса](https://uk.wikipedia.org/wiki/Формула_гаверсинуса). Вона дозволяє легко та швидко отримувати результати для різних точок, використовуючи географічні координати.

2. **Бібліотека `geopy`**  
   Бібліотека `geopy` надає можливість використання різних методів обчислення відстаней між точками на поверхні Землі, включаючи формулу гаверсинуса та інші методи. Вона також дозволяє використовувати різні моделі форми Землі для більш точних обчислень.

3. **Власна реалізація формули гаверсинуса**  
   Для порівняння була реалізована власна функція, що обчислює відстань між двома точками на поверхні Землі за допомогою формули гаверсинуса. Це дозволяє отримати результати без використання сторонніх бібліотек та зберегти простоту й прозорість коду.

### Формула гаверсинуса

[Формула гаверсинуса](https://uk.wikipedia.org/wiki/Формула_гаверсинуса) використовується для обчислення відстані між двома точками на поверхні сфери на основі їхніх географічних координат. Формула виглядає так:

$$ 
d = 2r \cdot \arcsin\left(\sqrt{\sin^2\left(\frac{\Delta \phi}{2}\right) + \cos(\phi_1) \cdot \cos(\phi_2) \cdot \sin^2\left(\frac{\Delta \lambda}{2}\right)}\right) 
$$

де:
- $\phi_1, \phi_2$ — широти двох точок у радіанах,
- $\Delta \phi = \phi_2 - \phi_1$ — різниця широт,
- $\Delta \lambda = \lambda_2 - \lambda_1$ — різниця довгот,
- $r$ — радіус Землі (приблизно 6371 км).

Ця формула добре підходить для обчислення відстаней на сферичній моделі Землі, що є досить точним у більшості випадків.

## Результати порівняння
- **Найбільша розбіжність результатів власної імплементації з `haversine` (м)**: 1.34 (Добротвірська ТЕС <> Курахівська ТЕС).  
  **У відсотках**: 0.000138% (Криворізька ТЕС <> Придніпровська ТЕС).

- **Найбільша різниця з `geopy` (м)**: 2799.20 (Добротвірська ТЕС <> Курахівська ТЕС).  
  **У відсотках**: 0.31% (Харківська ТЕЦ-5 <> Хмельницька АЕС).

Різниця відстаней між станціями, обчисленими за допомогою власної реалізації формули гаверсинуса та бібліотеки `haversine`, була незначною — менше 1.34 метрів для обраних точок. При цьому у відсотках ця різниця становила сталу величину, менше 0.000138% для всіх пар станцій.

Розбіжності з результатами `geopy` виявилися значно більшими — для деяких пар станцій вони досягали майже трьох кілометрів (2799 метрів для відстані між Добротвірською та Курахівською ТЕС). Ці розбіжності можуть бути пояснені наступними факторами:

1. **Різні моделі форми Землі**  
   Власна реалізація та бібліотека `haversine` використовують сферичну модель Землі з фіксованим радіусом (6371 км). Натомість `geopy` використовує еліпсоїдальну модель, яка враховує більш складну форму Землі, що дозволяє отримати більш точні результати для обчислення геодезичних відстаней.

2. **Точність обчислень**  
   Формула гаверсинуса є наближеною і добре працює для більшості практичних задач. Проте, для дуже точних обчислень, зокрема на великих відстанях, можуть бути відчутні різниці порівняно з методами, що враховують еліпсоїдальну форму Землі, як у `geopy`.

3. **Облік додаткових факторів**  
   Бібліотека `geopy` може враховувати додаткові фактори, такі як висота над рівнем моря та локальні особливості рельєфу, що також можуть впливати на результати обчислень.

## Висновки

Результати показують, що для завдань, де важлива висока точність географічних обчислень, варто використовувати методи, що враховують реальну форму Землі (як, наприклад, у бібліотеці `geopy`). Однак, для більшості практичних задач, особливо коли важлива простота та самодостатність коду, використання формули гаверсинуса є цілком прийнятним і надійним рішенням.

Для швидких обчислень на невеликих відстанях можна використовувати бібліотеку `haversine`, яка надає простий та зручний інтерфейс для роботи з географічними координатами.

Проте для обчислення ваги ребер у цій роботі було вирішено використовувати власну реалізацію формули гаверсинуса замість сторонніх бібліотек (`geopy` та `haversine`). Основні причини цього вибору:

1. **Простота та самодостатність коду**  
   Використання стандартних бібліотек дозволяє уникнути додаткових залежностей, що спрощує встановлення та використання скрипту в різних середовищах.

2. **Прозорість обчислень**  
   Власна реалізація дає можливість краще розуміти та контролювати процес обчислень, що важливо для точності та відстеження потенційних похибок.

3. **Зменшення потенційних конфліктів**  
   Відсутність сторонніх бібліотек зменшує ризик конфліктів версій та проблем сумісності, що може виникати при розгортанні коду в різних системах.


```python
from math import atan2, cos, radians, sin, sqrt

import matplotlib.pyplot as plt
import pandas as pd
from geopy.distance import geodesic
from haversine import haversine as haversine_lib


# Функція для обчислення відстані між двома точками за географічними координатами
def haversine_own(lat1, lon1, lat2, lon2):
    R = 6371.0  # Радіус Землі в кілометрах
    dlat = radians(lat2 - lat1)
    dlon = radians(lon2 - lon1)
    a = (
        sin(dlat / 2) ** 2
        + cos(radians(lat1)) * cos(radians(lat2)) * sin(dlon / 2) ** 2
    )
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    distance = R * c
    return distance


# Дані про станції
stations = {
    "Бурштинська ТЕС": (49.2066, 24.6255, 2300),
    "Добротвірська ТЕС": (50.0046, 24.2952, 700),
    "Криворізька ТЕС": (47.9142, 33.3233, 2820),
    "Курахівська ТЕС": (47.9780, 37.2827, 1500),
    "Ладижинська ТЕС": (48.6536, 29.2692, 1800),
    "Придніпровська ТЕС": (48.4472, 34.9943, 1765),
    "Південноукраїнська АЕС": (47.8148, 31.2065, 3000),
    "Рівненська АЕС": (51.3272, 25.0928, 2835),
    "Трипільська ТЕС": (50.1023, 30.7673, 1800),
    "Харківська ТЕЦ-5": (50.0056, 36.2292, 540),
    "Хмельницька АЕС": (50.3016, 26.6419, 2000),
}

# Всі можливі ребра (використовується для тестування)
edges = [
    ("Бурштинська ТЕС", "Добротвірська ТЕС"),
    ("Бурштинська ТЕС", "Криворізька ТЕС"),
    ("Бурштинська ТЕС", "Курахівська ТЕС"),
    ("Бурштинська ТЕС", "Ладижинська ТЕС"),
    ("Бурштинська ТЕС", "Придніпровська ТЕС"),
    ("Бурштинська ТЕС", "Південноукраїнська АЕС"),
    ("Бурштинська ТЕС", "Рівненська АЕС"),
    ("Бурштинська ТЕС", "Трипільська ТЕС"),
    ("Бурштинська ТЕС", "Харківська ТЕЦ-5"),
    ("Бурштинська ТЕС", "Хмельницька АЕС"),
    ("Добротвірська ТЕС", "Криворізька ТЕС"),
    ("Добротвірська ТЕС", "Курахівська ТЕС"),
    ("Добротвірська ТЕС", "Ладижинська ТЕС"),
    ("Добротвірська ТЕС", "Придніпровська ТЕС"),
    ("Добротвірська ТЕС", "Південноукраїнська АЕС"),
    ("Добротвірська ТЕС", "Рівненська АЕС"),
    ("Добротвірська ТЕС", "Трипільська ТЕС"),
    ("Добротвірська ТЕС", "Харківська ТЕЦ-5"),
    ("Добротвірська ТЕС", "Хмельницька АЕС"),
    ("Криворізька ТЕС", "Курахівська ТЕС"),
    ("Криворізька ТЕС", "Ладижинська ТЕС"),
    ("Криворізька ТЕС", "Придніпровська ТЕС"),
    ("Криворізька ТЕС", "Південноукраїнська АЕС"),
    ("Криворізька ТЕС", "Рівненська АЕС"),
    ("Криворізька ТЕС", "Трипільська ТЕС"),
    ("Криворізька ТЕС", "Харківська ТЕЦ-5"),
    ("Криворізька ТЕС", "Хмельницька АЕС"),
    ("Курахівська ТЕС", "Ладижинська ТЕС"),
    ("Курахівська ТЕС", "Придніпровська ТЕС"),
    ("Курахівська ТЕС", "Південноукраїнська АЕС"),
    ("Курахівська ТЕС", "Рівненська АЕС"),
    ("Курахівська ТЕС", "Трипільська ТЕС"),
    ("Курахівська ТЕС", "Харківська ТЕЦ-5"),
    ("Курахівська ТЕС", "Хмельницька АЕС"),
    ("Ладижинська ТЕС", "Придніпровська ТЕС"),
    ("Ладижинська ТЕС", "Південноукраїнська АЕС"),
    ("Ладижинська ТЕС", "Рівненська АЕС"),
    ("Ладижинська ТЕС", "Трипільська ТЕС"),
    ("Ладижинська ТЕС", "Харківська ТЕЦ-5"),
    ("Ладижинська ТЕС", "Хмельницька АЕС"),
    ("Придніпровська ТЕС", "Південноукраїнська АЕС"),
    ("Придніпровська ТЕС", "Рівненська АЕС"),
    ("Придніпровська ТЕС", "Трипільська ТЕС"),
    ("Придніпровська ТЕС", "Харківська ТЕЦ-5"),
    ("Придніпровська ТЕС", "Хмельницька АЕС"),
    ("Південноукраїнська АЕС", "Рівненська АЕС"),
    ("Південноукраїнська АЕС", "Трипільська ТЕС"),
    ("Південноукраїнська АЕС", "Харківська ТЕЦ-5"),
    ("Південноукраїнська АЕС", "Хмельницька АЕС"),
    ("Рівненська АЕС", "Трипільська ТЕС"),
    ("Рівненська АЕС", "Харківська ТЕЦ-5"),
    ("Рівненська АЕС", "Хмельницька АЕС"),
    ("Трипільська ТЕС", "Харківська ТЕЦ-5"),
    ("Трипільська ТЕС", "Хмельницька АЕС"),
    ("Харківська ТЕЦ-5", "Хмельницька АЕС"),
]

# Обчислення відстаней між станціями за трьома методами
results = []

for edge in edges:
    station1, station2 = edge
    lat1, lon1, _ = stations[station1]
    lat2, lon2, _ = stations[station2]
    distance_own = haversine_own(lat1, lon1, lat2, lon2)
    distance_lib = haversine_lib((lat1, lon1), (lat2, lon2))
    distance_geopy = geodesic((lat1, lon1), (lat2, lon2)).km
    results.append((station1, station2, distance_own, distance_lib, distance_geopy))

# Створення DataFrame для зручності аналізу
df = pd.DataFrame(
    results,
    columns=["Станція 1", "Станція 2", "Власна реалізація", "'haversine'", "'geopy'"],
)

# Додавання нумерації пар станцій
df.index = df.index + 1
df.index.name = "Пара"

# Обчислення відхилень у метрах та відсотках
# Відхилення власної реалізації з бібліотечним 'haversine'
df["Різниця з 'haversine' (м)"] = (
    abs(df["Власна реалізація"] - df["'haversine'"]) * 1000
)
df["haversine (%)"] = (
    df["Різниця з 'haversine' (м)"] / (df["'haversine'"] * 1000)
) * 100

# Відхилення результатів власної реалізації гаверсинусу від бібліотеки 'geopy'
df["Різниця з 'geopy' (м)"] = abs(df["Власна реалізація"] - df["'geopy'"]) * 1000
df["geopy (%)"] = (df["Різниця з 'geopy' (м)"] / (df["'geopy'"] * 1000)) * 100

# Розбіжнічть результатів у метрах з 'haversine'
max_deviation_haversine = df["Різниця з 'haversine' (м)"].max()
max_deviation_pair_haversine = df.loc[
    df["Різниця з 'haversine' (м)"].idxmax(), ["Станція 1", "Станція 2"]
]

# Розбіжнічть результатів у відсотках з 'haversine'
max_deviation_haversine_percent = df["haversine (%)"].max()
max_deviation_pair_haversine_percent = df.loc[
    df["haversine (%)"].idxmax(), ["Станція 1", "Станція 2"]
]

# Розбіжнічть результатів у метрах з 'geopy'
max_deviation_geopy = df["Різниця з 'geopy' (м)"].max()
max_deviation_pair_geopy = df.loc[
    df["Різниця з 'geopy' (м)"].idxmax(), ["Станція 1", "Станція 2"]
]

# Розбіжнічть результатів у відсотках з 'geopy'
max_deviation_geopy_percent = df["geopy (%)"].max()
max_deviation_pair_geopy_percent = df.loc[
    df["geopy (%)"].idxmax(), ["Станція 1", "Станція 2"]
]

# Виведення результатів
print("Результати обчислень відстаней між станціями за трьома методами:")

print(
    f"\nНайбільша різниця з 'haversine' (м): {max_deviation_haversine:.2f}\nміж станціями {max_deviation_pair_haversine['Станція 1']} і {max_deviation_pair_haversine['Станція 2']}"
)
print(
    f"\nНайбільша різниця з 'haversine' (%): {max_deviation_haversine_percent:.2f}\nміж станціями {max_deviation_pair_haversine_percent['Станція 1']} і {max_deviation_pair_haversine_percent['Станція 2']}"
)

print(
    f"\nНайбільша різниця з 'geopy' (м): {max_deviation_geopy:.2f}\nміж станціями {max_deviation_pair_geopy['Станція 1']} і {max_deviation_pair_geopy['Станція 2']}"
)

print(
    f"\nНайбільша різниця з 'geopy' (%): {max_deviation_geopy_percent:.2f}\nміж станціями {max_deviation_pair_geopy_percent['Станція 1']} і {max_deviation_pair_geopy_percent['Станція 2']}"
)

# Вертикальний відступ перед таблицею
print()

# Виведення результатів у таблиці
pd.set_option("display.max_rows", None)  # Виведення всіх рядків
print(df.to_string())


# Побудова графіку для порівняння
plt.figure(figsize=(15,6))
plt.plot(df.index, df["Власна реалізація"], label="Власна реалізація", marker="s")
plt.plot(df.index, df["'haversine'"], label="Бібліотека haversine", marker="o")
plt.plot(df.index, df["'geopy'"], label="Бібліотека geopy", marker="x")
plt.xlabel("Пари станцій")
plt.ylabel("Відстань (км)")
plt.title("Порівняння відстаней між станціями за трьома методами")
plt.legend()
plt.grid()
plt.show()

# Побудова графіку для відхилень у метрах з 'haversine'
plt.figure(figsize=(15,6))
plt.bar(df.index, df["Різниця з 'haversine' (м)"])
plt.xlabel("Пари станцій")
plt.ylabel("Відхилення (м)")
plt.title(
    "Відхилення між відстанями за власною реалізацією та бібліотекою haversine (в метрах)"
)
plt.grid()
plt.show()

# Побудова графіку для відхилень у відсотках з 'haversine'
plt.figure(figsize=(15,6))
plt.bar(df.index, df["haversine (%)"])
plt.xlabel("Пари станцій")
plt.ylabel("Відхилення (%)")
plt.title(
    "Відхилення між відстанями за власною реалізацією та бібліотекою haversine (у відсотках)"
)
plt.grid()
plt.show()

# Побудова графіку для відхилень у метрах з 'geopy' у метрах
plt.figure(figsize=(15,6))
plt.bar(df.index, df["Різниця з 'geopy' (м)"])
plt.xlabel("Пари станцій")
plt.ylabel("Відхилення (м)")
plt.title(
    "Відхилення між відстанями за власною реалізацією та бібліотекою geopy (в метрах)"
)
plt.grid()
plt.show()

# Побудова графіку для відхилень у відсотках з 'geopy'
plt.figure(figsize=(15,6))
plt.bar(df.index, df["geopy (%)"])
plt.xlabel("Пари станцій")
plt.ylabel("Відхилення (%)")
plt.title(
    "Відхилення між відстанями за власною реалізацією та бібліотекою geopy (у відсотках)"
)
plt.grid()
plt.show()
```
